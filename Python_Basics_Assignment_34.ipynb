## Question 2
**Explain the difference between List and Tuple in Python.**

### Answer 2

| Feature | List | Tuple |
|--------|------|-------|
| **Mutability** | Mutable (can be changed after creation) | Immutable (cannot be changed after creation) |
| **Syntax** | Defined using square brackets `[ ]` | Defined using parentheses `( )` |
| **Performance** | Slightly slower due to mutability | Faster than lists |
| **Use Case** | Used when data needs to be modified | Used for fixed/constant data |
| **Methods** | Has more built-in methods (append, remove, etc.) | Has fewer methods (count, index) |

````python
# Demonstrating List vs Tuple

# List - Mutable
my_list = [1, 2, 3, 4]
print("Original List:", my_list)
my_list[0] = 10  # Modifying a list
print("Modified List:", my_list)

# Tuple - Immutable
my_tuple = (1, 2, 3, 4)
print("\nTuple:", my_tuple)
# my_tuple[0] = 10  # This would raise a TypeError
print("Tuples cannot be modified after creation.")
````

## Question 3
**What is a function in Python? Why are functions useful?**

### Answer 3

A **function** in Python is a named block of reusable code that performs a specific task. It is defined using the `def` keyword, followed by the function name and parentheses.

**Syntax:**
```python
def function_name(parameters):
    # code block
    return value
```

**Why are functions useful?**

1. **Reusability**: Write code once and call it multiple times anywhere in the program, reducing redundancy.
2. **Modularity**: Functions break a large program into smaller, manageable pieces, making it easier to read and maintain.
3. **Avoid Repetition (DRY Principle)**: Instead of writing the same logic again and again, a function allows you to define it once and reuse it — following the "Don't Repeat Yourself" principle.
4. **Easy Debugging**: Errors can be isolated and fixed within a specific function without affecting the rest of the code.

````python
# Example of a simple function
def greet_user(name):
    return f"Hello, {name}! Welcome to Python."

# Calling the function multiple times
print(greet_user("Alice"))
print(greet_user("Bob"))
````

## Question 6
**Write a program to print numbers from 1 to 10 using a loop.**

### Answer 6

````python
# Program to print numbers from 1 to 10 using a for loop

print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i, end=" ")

print()  # newline

# Using while loop (alternative approach)
print("\nUsing while loop:")
i = 1
while i <= 10:
    print(i, end=" ")
    i += 1
````

## Question 7
**Create a list of five numbers and print the maximum number from the list.**

### Answer 7

````python
# Create a list of five numbers and find the maximum

numbers = [34, 78, 12, 56, 90]
print("List of numbers:", numbers)

# Using built-in max() function
maximum = max(numbers)
print(f"Maximum number in the list: {maximum}")

# Alternative: Finding max manually without built-in function
max_num = numbers[0]
for num in numbers:
    if num > max_num:
        max_num = num
print(f"Maximum number (manual method): {max_num}")
````

## Question 8
**Write a Python program to remove duplicate values from a list using a set.**

### Answer 8

````python
# Program to remove duplicate values from a list using a set

original_list = [1, 2, 3, 2, 4, 3, 5, 1, 6]
print("Original List:", original_list)

# Convert list to set (removes duplicates), then back to list
unique_list = list(set(original_list))
print("List after removing duplicates:", sorted(unique_list))  # sorted for clean output

print(f"\nOriginal length: {len(original_list)}")
print(f"Length after removing duplicates: {len(unique_list)}")
````

**Write a function that returns the square of a number.**

### Answer 9

````python
# Function to return the square of a number

def square(number):
    """Returns the square of the given number."""
    return number ** 2

# Testing the function
print(f"Square of 5 = {square(5)}")
print(f"Square of 9 = {square(9)}")
print(f"Square of 12 = {square(12)}")

# Taking input from user
num = int(input("\nEnter a number to find its square: "))
print(f"Square of {num} = {square(num)}")
````

## Question 10
**Write a Python program to count how many times a number appears in a list.**

**Example List:** `[2, 3, 4, 2, 5, 2]`

### Answer 10

````python
# Program to count how many times a number appears in a list

my_list = [2, 3, 4, 2, 5, 2]
print("List:", my_list)

# Using list's built-in count() method
number_to_count = 2
count = my_list.count(number_to_count)
print(f"\nThe number {number_to_count} appears {count} time(s) in the list.")

# Count occurrences of all elements using a loop
print("\nCount of all elements:")
unique_elements = set(my_list)
for element in sorted(unique_elements):
    print(f"  {element} appears {my_list.count(element)} time(s)")

# Alternative: Using a counter variable manually
target = int(input("\nEnter a number to count in the list: "))
counter = 0
for item in my_list:
    if item == target:
        counter += 1
print(f"{target} appears {counter} time(s) in the list.")
````
